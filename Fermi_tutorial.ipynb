```python
from PulsarPhaseogram import *
from lstchain.io.io import dl2_params_lstcam_key,dl2_params_src_dep_lstcam_key
from astropy.io import fits
import pandas as pd
import matplotlib.pylab as plt
from IPython.display import Image
import numpy as np
```

# Creating the PulsarAnalysis object and settings

```python
h=PulsarAnalysis()
```

```python
h.setFermiInputFile('./PulsarPhaseogram/Tutorial/example_files/merged2_pulsar.fits')
```

```python
h.setBackgroundLimits([0.52,0.87])
h.setPeaklimits(P1_limits=[0,0.026,0.983,1],P2_limits=[0.377,0.422],P3_limits=None)
```

```python
h.setBinning(50,xmin=0,xmax=1)
```

```python
h.setTimeInterval(tint=3600*24*30)
```

```python
h.setFittingParams(model='dgaussian',binned=False)
```

```python
h.setEnergybinning(np.logspace(np.log10(0.1),np.log10(10), 5))
```

# Run the code

```python
h.run()
```
